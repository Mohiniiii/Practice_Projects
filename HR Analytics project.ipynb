The above dataset is related to the attrition rate and the factors affecting it in any organization. Attrition in human resources refers to the gradual loss of employees overtime. # About some important Columns :

1. Age : The first column is about the age of the employees working in any organization.

2. Attrition :  a gradual but deliberate reduction in staff numbers that occurs as employees retire or resign and are not replaced.Yes means that there is some kind of attrition going on. No means attrition is not there.

3. Business Travel: This means that employees travel due to business purposes. This has been categorised into Travel rarely, Travel Frequently and Non-Travel categories.

4. Daily Rate : refers to the rate of attrition occurring .

5. Department : tells us about the respective departments of the employees.

6. Distance from home: tells about the distance of the employees from their work location in km.

7. The column Education has 5 categories : 1,2,3,4 and 5 . 1 stands for Below college, 2 stands for college , 3 stands for Bachelor, 4 stands for Master and 5 stands for Doctor.

8. Education Field column tells about the field or area of one's education.

9. Environment Satisfaction : this has been categorised into 4 categories .1 stands for Low , 2 stands for medium, 3 stands for high and 4 stands for very high.

10. job Involvement : This column has categorised the involvement into 4 categories . 1- Low, 2-Medium, 3-High, 4-very High.

11. Performance Rating : This column categorises the performance into 4 categories : 1-Low, 2-Good, 3-Excellent, 4-Outstanding.

12. Relationship Satisfaction : This column categorises the satisfaction into 4 categories . 1- Low, 2-Medium , 3-High, 4-Very High.

13. Work Life Balance :This column categorises this factor into 4 categories . 1-Bad, 2-Good, 3-Better, 4-Best.

Creating age -ranges :

```python
agerange = []
for age in dt["Age"]:
    if age >= 18 and age < 24:
        agerange.append("18-24")
    elif age >= 25 and age < 31:
        agerange.append("25-31")
    elif age >= 32 and age < 38:
        agerange.append("32-38")
    elif age >= 39 and age < 45:
        agerange.append("39-45")
    elif age >= 46 and age < 52:
        agerange.append("46-52")
    elif age >= 53 and age < 59:
        agerange.append("53-59")
    else:
        agerange.append("60-66")
       
dt["AgeRange"] = agerange
```

# Training the dataset :

```python
x = dt.drop(['Attrition', 'AgeRange', 'DistanceFromHome', 'NumCompaniesWorked', 'WorkLifeBalance','BusinessTravel','JobRole','Over18'], axis=1)
```

```python
y = dt[['Attrition']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x
```

```python
x.columns
```

# Scaling the data :

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

# Applying Logistic Regression to the model :

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr = lr.fit(x_train, y_train)

# check the accuracy on the training set
print('Accuracy =', lr.score(x_train, y_train)*100,'%');
```

Checking accuracy on test data :

```python
x_test_prediction=lr.predict(x_test)
```

```python
x_test_prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
#now again we will compare as above.as our machine has not seen the test data.
testing_data_accuracy=accuracy_score(y_test,x_test_prediction)
```

```python
testing_data_accuracy
```

# Using Grid Search CV for data tuning :

Using Ridge Model :

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
#prepeare a range of alpha values to test
alphavalue={'alpha': [1,0.1,0.01,0.001,0.0001,0]}
#create and fit a ridge regression model , testing each alpha
model= Ridge()
grid=GridSearchCV(estimator=model,param_grid=alphavalue)

grid.fit(x, y)
print(grid)
#summarize the result of grid search
print(grid.best_score_)
print(grid.best_estimator_.alpha)
print(grid.best_params_)
```

```python
from sklearn.linear_model import Lasso , Ridge
rd=Ridge(alpha=1)
rd.fit(x,y)
print(rd.coef_)
```

Using Decision Tree Classifier :

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
grid_param={'criterion': ['gini','entropy']}

#gridk={'kernel':['linear','rbf','poly']}

gd_sr=GridSearchCV(estimator=dtc,param_grid=grid_param,scoring='accuracy',cv=3)

gd_sr.fit(x,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(round(best_result,2))
```

```python
dtc=DecisionTreeClassifier(criterion='gini')
dtc.fit(x,y)
```
